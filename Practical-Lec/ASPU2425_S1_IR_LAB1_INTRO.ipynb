## pandas

[pandas](http://pandas.pydata.org/pandas-docs/stable/) is a Python library for data wrangling and analysis. It provides:

DataFrame: a table, similar to an R DataFrame that holds any structured data

Every column can have its own data type (strings, dates, floats,…)

A great range of methods to apply to this table (sorting, querying, joining,…)

Imports data from a wide range of data formats (CSV, Excel) and databases (e.g. SQL)


```python
# %pip install pandas
```

```python
import pandas as pd
import numpy as np
```

Series

A one-dimensional array of data (of any numpy type), with indexed values.

It can be created by passing a Python list or dict, a numpy array, a csv file,…


```python
pd.Series([1, 3, np.nan])  # Default integers are integers
pd.Series([1, 3, 5], index=["a", "b", "c"])
pd.Series(
    {"a": 1, "b": 2, "c": 3}
)  # when given a dict, the keys will be used for the index
pd.Series(
    {"a": 1, "b": 2, "c": 3}, index=["b", "c", "d"]
)  # this will try to match labels with keys
```

Functions like a numpy array, however with index labels as indices


```python
a = pd.Series({"a": 1, "b": 2, "c": 3})
print(a)
print(a["b"])  # Retrieves a value
print(a[["a", "b"]])  # and can also be sliced
```

numpy array operations on Series preserve the index value


```python
print(a)
print(a[a > 1])
print(a * 2)
print(np.sqrt(a))
```

**DataFrame**


```python
data = {
    "state": ["Ohio", "Ohio", "Nevada", "Nevada"],
    "year": [2000, 2001, 2001, 2002],
    "pop": [1.5, 1.7, 2.4, 2.9],
}
pd.DataFrame(data)
pd.DataFrame(data, columns=["year", "state", "pop", "color"])  # Will match indices
```

It can be composed with a numpy array and row and column indices


```python
dates = pd.date_range("20130101", periods=4)
df = pd.DataFrame(np.random.randn(4, 4), index=dates, columns=list("ABCD"))
df
```

```python
df.index
df.columns
df.values
```

DataFrames can easily read/write data from/to files

- read_csv(source): load CSV data from file or url
- read_table(source, sep=','): load delimited data with separator
- df.to_csv(target): writes the DataFrame to a file


```python
df.to_csv("data.csv", index=False)  # Don't export the row index
dfs = pd.read_csv("data.csv")
dfs
dfs.at[0, "A"] = 10  # Set value in row 0, column 'A' to '10'
dfs.to_csv("data.csv", index=False)
```

**Indexing and slicing**


```python
print(df["A"])  # Get single column by label
print(df.A)  # Shorthand
```

```python
print(df[0:2])  # Get rows by index number
print(df.iloc[0:2, 0:2])  # Get rows and columns by index number
print(df["20130102":"20130103"])  # or row label
print(df.loc["20130101":"20130103", ["A", "B"]])  # or row and column label
```

query() retrieves data matching a boolean expression


```python
print(df)
print(df.query("A > -0.4"))  # Identical to df[df.A > 0.4]
print(df.query("A > B"))  # Identical to df[df.A > df.B]
```

**Operations**

DataFrames offer a wide range of operations: max, mean, min, sum, std,…


```python
print(df.mean())  # Mean of all values per column
print(df.mean(axis=1))  # Other axis: means per row
```

```python
# All of numpy’s universal functions also work with dataframes
np.abs(df)
```

```python
# Other (custom) functions can be applied with apply(funct)
print(df)
print(df.apply(np.max))
print(df.apply(lambda x: x.max() - x.min()))
```

```python
# Data can be aggregated with groupby()

df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar"],
        "B": ["one", "one", "two", "three"],
        "C": np.random.randn(4),
        "D": np.random.randn(4),
    }
)
print(df)
print(df.groupby("A").sum())
print(df.groupby(["A", "B"]).sum())
```

```python
# Merge: combine two dataframes based on common keys

df1 = pd.DataFrame({"key": ["b", "b", "a"], "data1": range(3)})
df2 = pd.DataFrame({"key": ["a", "b"], "data2": range(2)})
print(df1)
print(df2)
print(pd.merge(df1, df2))
```

```python
# Append: append one dataframe to another

df = pd.DataFrame(np.random.randn(2, 4))
print(df)
s = pd.DataFrame(np.random.randn(1, 4))
print(s)
df = pd.concat([df, s], ignore_index=True)
```

```python
# Count and Remove duplicates

df = pd.DataFrame({"k1": ["one"] * 3, "k2": [1, 1, 2]})
print(df, "\n")
print(df.value_counts(), "\n")
print(df.drop_duplicates(), "\n")
```

```python
# Replace values

df = pd.DataFrame(
    {"k1": [1, -1], "k2": [-1, 2]}
)  # Say that -1 is a sentinel for missing data
df
df.replace(-1, np.nan)
```
